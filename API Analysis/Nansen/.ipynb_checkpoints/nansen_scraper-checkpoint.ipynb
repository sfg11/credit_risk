```python
#go to site 
def go_to_site(site,**kwargs):
    #site = "https://etherscan.io/enslookup?ps={rows}&p={pg}".format(**{'rows':rows,'pg':pg})
    #site = "https://pro.nansen.ai/"
    site = site.format(**kwargs)
    return site
```

```python
def secs():
    return np.random.randint(5,10)
```

```python
#go to login site
site= "https://pro.nansen.ai/"
driver.get(site)
```

```python
username_input = '//*[@id="formBasicEmail"]'
password_input = '//*[@id="formBasicPassword"]'
login_submit = '//*[@id="root"]/div/div/div/div/div[1]/form/button'
fields  = {"username":  'udai@teller.finance' , "password": 'TellerFinance' }
driver.find_element_by_xpath(username_input).send_keys(fields["username"])
driver.find_element_by_xpath(password_input).send_keys(fields["password"])
driver.find_element_by_xpath(login_submit).click()
```

```python
fl='covalent_balance_snapshot_as_of_5_28.gzip'
df_res=pd.read_parquet(fl)
address = df_res['address'].values
```

```python
wallet_info = {} #address, esn,
index = 0
failed_wallet = []
for ix,addr in enumerate(address[351:1000]):
    print('index is at ... ', ix)
    site= 'https://pro.nansen.ai/wallet-profiler?address={address}'
    dic = {'address':addr} 
    site = go_to_site(site,**dic)
    driver.get(site)
    time.sleep(3*secs())
    res = driver.find_element_by_xpath('//*[@id="root"]/div/div/div/main/div/div[1]/div[1]/article')
    print(' profile for wallet ... ', addr, '\n', res.text )
    cnt = 0
    
    while 'Loading' in res.text  and cnt <= 5:
        cnt = cnt + 1
        time.sleep(secs())
        res = driver.find_element_by_xpath('//*[@id="root"]/div/div/div/main/div/div[1]/div[1]/article')
        print(' profile for wallet ... ', addr, '\n', res.text )
        if cnt >  3:
            driver.get(site)
            time.sleep(5*secs()) # try res again, otherwise while  statement will fail with stale object error
            res = driver.find_element_by_xpath('//*[@id="root"]/div/div/div/main/div/div[1]/div[1]/article')

            
    if cnt == 5 and 'Loading' in res.text:
        failed_wallet.append(dic['address'])
        
    else:
    
        if 'Show More' in res.text:
            driver.find_element_by_xpath('//*[@id="root"]/div/div/div/main/div/div[1]/div[1]/article/div[2]/button').click()
            res = driver.find_element_by_xpath('//*[@id="portal-root"]/div[2]/div/div/div')


        if wallet_info == {}:
            wallet_info['index'] = [ix]
            wallet_info['address'] = [dic['address']] #assuming all wallet addresses start with 0x
            wallet_info['raw_string'] = [res.text]
        else:
            wallet_info['index'].append(wallet_info['index'][index] + 1)
            wallet_info['address'].append( dic['address'] )
            wallet_info['raw_string'].append( res.text )
            index = index + 1

    if ix!=0 and np.mod(ix,9) == 0: #save every 10 element
        df_res = pd.DataFrame.from_dict(wallet_info)
        name= 'nansen_wallet_profiler'+ '\\' +'nansen_wallet_index_' +  str(ix) +'.gzip'
        df_res.to_parquet(name, compression='gzip')
        wallet_info = {}
        index = 0
        se = 50*secs() # at least a 4 minute break
        print('being nice and waiting...', se, ' seconds')
        time.sleep(se)
    
    if ix!=0 and np.mod(ix,50) == 0:
        se = 100*secs()
        print('being EXTRA nice and waiting...', se, ' extra seconds')
        time.sleep(se)
```
